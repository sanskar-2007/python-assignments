```python
# Assignment No: 04

# 1) WAP to display even numbers from 1 to 10
a = 2
while a <= 10:
    print(a)
    a += 2


# 2) WAP to add odd numbers from 1 to 10
sum = 0
for i in range(1, 11):
    if i % 2 != 0:
        sum += i

print("Sum of odd numbers:", sum)


# 3) WAP to print Fibonacci series between 0 to 50
a = 0
b = 1

print("Fibonacci series up to 50:")
while a <= 50:
    print(a)
    a, b = b, a + b


# 4) WAP to remove characters with odd index values from a string
string = input("Enter a string: ")

result = ""
for i in range(len(string)):
    if i % 2 == 0:   # keep even index characters
        result += string[i]

print("String after removing odd index characters:", result)
```
